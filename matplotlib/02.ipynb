```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
y
```

```python
fig = plt.figure(dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(2,8)
axes1.set_ylim(2,800)
axes1.set_xlabel("A-axis")
axes1.set_ylabel('B-axis')
axes1.set_title('expo graph')

axes2 = fig.add_axes([1,0,1,1])
axes2.plot(x,y)
axes2.set_xlim(1,8)
axes2.set_ylim(4,10)
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
axes2.set_title('lin graph')

fig.savefig('fig1.png',bbox_inches='tight')
plt.show()
```
