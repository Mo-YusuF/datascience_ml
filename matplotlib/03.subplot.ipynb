```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(1,10,11)
b = 4**a
```

```python
x = np.arange(0,11)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
# axes[0][1].plot(x,y)
axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,dpi=200)
# axes[0][1].plot(x,y)
axes[0][0].plot(x,y)
axes[0][0].set_xlabel('x')
axes[0][0].set_ylabel('y')

axes[1][1].plot(a,b)
axes[1][1].set_title('expo-func')

fig.suptitle('Figure Level',fontsize=18)

fig.tight_layout()
# fig.subplots_adjust(wspace=0.5,hspace=0.4)
```
