```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,11)
y = 2*x
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#e5ed4a',label='x vs x',lw=3,ls='--')  #can assign any hex color

ax.plot(x,x+1,color='green',label='x vs x+1',ls=':')

ax.plot(x,x-1,color='purple',label='x vs x-1',ls='-.')

ax.legend()
```

```python
line = plt.plot(x,x,color='teal',lw=3)
line[0].set_dashes([1,2,3,5,10])
```

```python
#adding markers on line
ax = plt.plot(x,x,marker='o')

ax = plt.plot(x,x+1,marker='+',markersize=20)
```
